```python
import numpy as np
import matplotlib.pyplot as plt

import os, glob
%matplotlib inline
```

```python
def openraw(filename, shape=(1234, 1624), reshape=True):
    f = open(filename, 'rb')  #apre in binario
    rawdata = np.fromfile(f,dtype=np.uint16).astype(int)
    f.close
    if reshape:
        try:
            return rawdata.reshape(shape)
        except ValueError:
            print('Actual data length is %s, but %d is required for shape %d x %d'%(rawdata.shape, shape[0]*shape[1], shape[1], shape[0]))
            return rawdata
    else:
        return rawdata
```
