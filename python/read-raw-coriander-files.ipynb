```python
import numpy as np
import matplotlib.pyplot as plt

import os
%matplotlib inline
```

```python
def openraw(filename, reshape=True):
    f = open(filename, 'rb')  #apre in binario
    rawdata = np.fromfile(f,dtype=np.uint16).astype(int)
    # 'H' = uint16
    # lo trovo in np.typeDict
    #legge in array i dati formattati uint16 e li casta ad int
    ''' NB Ã¨ importantissimo castarli ad int:
        visto che sono numeri molto grandi, gli unsigned short non vanno bene (overflow) '''
    f.close
    if reshape:
        return rawdata.reshape((1234,1624))
    else:
        return rawdata
```
