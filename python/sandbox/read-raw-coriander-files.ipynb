```python
import numpy as np
import matplotlib.pyplot as plt

import os, glob
%matplotlib inline

def openraw(filename, shape=(1234, 1624), reshape=True):
    f = open(filename, 'rb')  #apre in binario
    raw = np.fromfile(f,dtype=np.uint16).astype(int)
    header = np.fromfile(f)
    f.close()
    rawdata = raw[-shape[0]*shape[1]:]
    if reshape:
        try:
            return rawdata.reshape(shape), header
        except ValueError:
            print('Actual data length is %s, but %d is required for shape %d x %d'%(rawdata.shape, shape[0]*shape[1], shape[1], shape[0]))
            return rawdata, header
    else:
        return rawdata, header
```
